#### 💻 [Code 单元格 1]：极其无脑的 10 行代码

*(王博士画外音：同学们，不要害怕，我们不搞复杂的推导，我们直接写一个最基础的 `for` 循环，看看这个方程跑起来是什么样子。)*

```python
import numpy as np
import matplotlib.pyplot as plt

# 极其无脑的 10 行核心代码：二次映射迭代
def simple_chaos(mu, x0=0.1, steps=100):
    x_history = [x0]
    x = x0
    for _ in range(steps):
        x = 1 - mu * x**2  # 就这一行核心公式！
        x_history.append(x)
    return x_history

# 让我们拧动“上帝旋钮” mu，看看会发生什么
plt.figure(figsize=(10, 3))
plt.plot(simple_chaos(mu=0.5), label="mu=0.5 (Dead/Stable)")
plt.plot(simple_chaos(mu=1.8), alpha=0.8, label="mu=1.8 (Crazy/Chaos)")
plt.legend()
plt.title("Time Series of the Quadratic Map")
plt.show()
```

#### 💻 [Code 单元格 2]：大自然的指纹（分岔图）

*(王博士画外音：只看时间序列太枯燥了。如果我们把这个“上帝旋钮”  从 0 一点点慢慢拧到 2，把系统稳定下来的点全都画在一张图上，会看到什么？我们再加几行代码。)*

```python
# 绘制著名的分岔图 (Bifurcation Diagram)
mu_values = np.linspace(0, 2, 1000)
x = np.ones_like(mu_values) * 0.1 # 初始值

plt.figure(figsize=(10, 6))
# 前 100 步作为热身（抛弃瞬态）
for _ in range(100):
    x = 1 - mu_values * x**2

# 后 100 步画出系统真正的“骨架”
for _ in range(100):
    x = 1 - mu_values * x**2
    plt.scatter(mu_values, x, s=0.1, color='black', alpha=0.5)

plt.title("Bifurcation Diagram: The Road to Chaos")
plt.xlabel("Control Parameter (mu)")
plt.ylabel("System State (x)")
plt.show()
```
