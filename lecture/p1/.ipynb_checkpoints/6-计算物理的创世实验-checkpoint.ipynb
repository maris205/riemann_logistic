### 🔴 小节 3.1 Jupyter Notebook 讲稿：非自治法则与算力巨兽

#### 📝 [Markdown 单元格 1]：写入创世法则

> 欢迎来到《创世计算器》第三季：封神局。
> 在过去的一百年里，数学家们一直在寻找一个能够完美生成黎曼零点的“静态算子”。但在上一集，我们证明了静态宇宙注定失败，因为黎曼零点服从外尔定律（Weyl's Law），系统必须随着尺度“老化”。
> 今天，我们将把这套老化法则，用一行代码直接刻进大自然的底层：
>
>
> * ：带合并相变点，宇宙最初的拓扑基石。

#### 💻 [Code 单元格 1]：召唤算力巨兽 (JIT 编译)

*(王博士画外音：同学们，看清楚了。普通的 Python 跑个一百万步就累得喘气。但今天我们要跑的是 （100 亿步）！在不租用 256 核超算集群的情况下，为了在视频里给大家演示，我们必须动用 `Numba` 黑科技，直接把 Python 编译成机器码！)*

```
import numpy as np
import scipy.sparse as sp
from scipy.sparse.linalg import eigs
from numba import njit
import time

# 终极参数设定（源自 2026-02-10 的极密实验数据）
U_C = 1.543689012
K_OPT = 12.32
STEPS = 10**10  # 100 亿步！尽显经费实力
N_BINS = 20000  # 极清分辨率：2万x2万矩阵

@njit
def run_non_autonomous_universe(steps, n_bins, u_c, k_opt):
    # 初始化统计数组
    counts = np.zeros(n_bins, dtype=np.float64)
    x = 0.5

    # 极速引擎启动！
    for i in range(1, steps + 1):
        # 核心法则：重整化流驱动的动态老化
        current_mu = u_c + k_opt / (np.log(i + 100)**2)
        x = 1.0 - current_mu * x**2

        # 越界保护
        if x > 1.0: x = 0.999
        elif x < -1.0: x = -0.999

        # 投射到网格
        bin_idx = int((x + 1.0) / 2.0 * (n_bins - 1))
        counts[bin_idx] += 1

    return counts

print(f"🚀 启动非自治宇宙模拟引擎...\n参数: U_c={U_C}, K={K_OPT}, 步数={STEPS}")
start_time = time.time()
# 【在视频里，这里可以剪辑一个极其酷炫的进度条，配合 CPU 满载的风扇声】
density = run_non_autonomous_universe(STEPS, N_BINS, U_C, K_OPT)
print(f"✅ 100 亿步演化完成！耗时: {time.time() - start_time:.2f} 秒")
```
