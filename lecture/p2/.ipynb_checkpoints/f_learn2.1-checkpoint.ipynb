```python
import numpy as np
import matplotlib.pyplot as plt

def hello_world_spectrum(r=2.0, N=1000):
    # =================================================================
    # 1. 划分空间：把连续的宇宙切成 1000 个离散的“方格”
    # =================================================================
    edges = np.linspace(-1, 1, N + 1)
    centers = (edges[:-1] + edges[1:]) / 2
    
    # 初始化转移矩阵 P (我们的“状态路由表”)
    P = np.zeros((N, N))
    
    # =================================================================
    # 2. 核心：最暴力的确定性映射 (Perron-Frobenius 算子的极简版)
    # =================================================================
    print("正在构建确定性路由表...")
    for j in range(N):
        # 提取当前方格 j 的中心点，代入动力学方程，算出下一步飞去哪
        x_next = 1 - r * (centers[j]**2)
        
        # 暴力定位：利用二分查找 (searchsorted)，看新位置落在了哪个方格 i
        # (因为 r=2.0 时 x_next 绝对在 [-1, 1] 内，所以直接减 1 即为索引)
        i = np.searchsorted(edges, x_next) - 1
        
        # 记账：从方格 j (出发地) 到方格 i (目的地) 的转移概率是 100%
        P[i, j] = 1.0  
        
    # =================================================================
    # 3. 抽取灵魂：计算特征值
    # =================================================================
    print("正在计算矩阵特征值...")
    eigenvalues = np.linalg.eigvals(P)
    
    # =================================================================
    # 4. 可视化：见证神奇的“单位圆”与“黑洞”
    # =================================================================
    plt.figure(figsize=(6, 6))
    
    # 画出作为参照的单位圆 (模长 |λ| = 1)
    theta = np.linspace(0, 2*np.pi, 100)
    plt.plot(np.cos(theta), np.sin(theta), 'k--', alpha=0.3, label="Unit Circle $(|\lambda|=1)$")
    
    # 画出真实算出的特征值散点
    plt.scatter(eigenvalues.real, eigenvalues.imag, s=10, c='red', alpha=0.6)
    
    plt.title(f"Hello World: Deterministic Spectrum ($r={r}$)", fontsize=14)
    plt.xlabel(r"Re($\lambda$)")
    plt.ylabel(r"Im($\lambda$)")
    plt.grid(True, alpha=0.2)
    plt.axis('equal')
    plt.legend()
    plt.show()

# 运行最经典的强混沌态 r=2.0
hello_world_spectrum(r=2.0, N=1000)
```
