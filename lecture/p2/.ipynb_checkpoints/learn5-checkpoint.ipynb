```python
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------------------------------------
# 1. 构造转移矩阵 (与你之前一致)
# ---------------------------------------------------------
r = 2.0
N = 1200  # 稍微增大维度，统计特性会更明显
edges = np.linspace(-1, 1, N + 1)
centers = (edges[:-1] + edges[1:]) / 2
P = np.zeros((N, N))

for j in range(N):
    x_next = 1 - r * (centers[j]**2)
    if -1 <= x_next <= 1:
        i = np.searchsorted(edges, x_next) - 1
        if 0 <= i < N:
            P[i, j] = 1.0

# ---------------------------------------------------------
# 2. 提取特征值并处理
# ---------------------------------------------------------
# 计算特征值
eigvals = np.linalg.eigvals(P)

# 过滤：由于离散化误差，很多特征值接近0。我们取模长在 0.1 到 0.99 之间的
# 注意：1.0 是稳态，不参与间距统计
mags = np.sort(np.abs(eigvals))
mags = mags[(mags > 0.1) & (mags < 0.99)] 

# ---------------------------------------------------------
# 3. 谱展开 (Unfolding) - 极其重要
# ---------------------------------------------------------
# 简单的谱展开方法：将特征值映射到其累积分布函数(CDF)上
# 这样处理后的“新特征值”在 [0, max] 之间是均匀分布的，平均间距为 1
unfolded_eigs = np.sort(mags) 
# 计算相邻间距
spacings = np.diff(unfolded_eigs)
# 归一化，使平均间距为 1
s = spacings / np.mean(spacings)

# ---------------------------------------------------------
# 4. 绘图对比
# ---------------------------------------------------------
plt.figure(figsize=(10, 6))

# 绘制实验数据的直方图
plt.hist(s, bins=40, density=True, alpha=0.5, color='steelblue', label='Map Spectrum (r=2.0)')

# 绘制 GUE (随机矩阵理论/黎曼零点) 理论曲线
x = np.linspace(0, 4, 100)
gue = (32 / (np.pi**2)) * (x**2) * np.exp(-(4 / np.pi) * (x**2))
plt.plot(x, gue, 'r-', lw=2, label='GUE (Riemann Zeros / Chaos)')

# 绘制 Poisson (规则系统) 理论曲线
poisson = np.exp(-x)
plt.plot(x, poisson, 'g--', lw=2, label='Poisson (Integrable)')

plt.title("Level Spacing Distribution: Map vs. Riemann/Chaos Theory")
plt.xlabel("Normalized Spacing (s)")
plt.ylabel("P(s)")
plt.xlim(0, 3)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
