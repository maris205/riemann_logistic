```python
import numpy as np
import os
import matplotlib.pyplot as plt

# 1. 准备真实零点数据
TRUE_GAMMAS = np.array([14.1347, 21.0220, 25.0108, 30.4248, 32.9350])
k_best_per_mode = []

# 2. 遍历所有扫描出来的结果文件
files = [f for f in os.listdir('.') if f.startswith('res_k_') and f.endswith('.npy')]
all_results = []

for f in files:
    # 从文件名提取 k 值
    k_val = float(f.split('_')[2])
    phases = np.load(f)
    
    # 模拟对齐逻辑：以第一个零点进行缩放校准
    scaling = TRUE_GAMMAS[0] / phases[0]
    sim_gammas = phases[:5] * scaling
    
    # 计算前 5 个模式的各自误差
    errors = sim_gammas - TRUE_GAMMAS
    all_results.append((k_val, errors))

# 3. 寻找令每个模式误差最小的 k 值
all_results.sort() # 按 k 排序
for i in range(5):
    # 找到第 i 个模式误差绝对值最小的索引
    best_idx = np.argmin([abs(res[1][i]) for res in all_results])
    best_k = all_results[best_idx][0]
    k_best_per_mode.append(best_k)
    print(f"模式 {i+1} 的最优锁定 k 值: {best_k:.4f}")
```
